```python
# Memory usage function
import os
def get_memory_usage(df,units='mb'):
    """returns memory size of dataframe in requested units"""
    memory = df.memory_usage().sum()

    if units.lower()=='mb':
        denom = 1e6
    elif units.lower()=='gb':
        denom = 1e9
    else:
        raise Exception('Units must be either "mb" or "gb"')
    val = memory/denom
    print(f"- Total Memory Usage = {val} {units.upper()}")
```

```python
# Getting the memory_usage of akas_df in small chunks
get_memory_usage(akas_df)
```

```python
# Filesize function
def get_filesize(fname, units='mb'):
    """Get size of file at given path in MB or GB"""
    if units.lower()=='mb':
        denom = 1e6
    elif units.lower()=='gb':
        denom = 1e9
    else:
        raise Exception('Units must be either "mb" or "gb"')

    import os
    size = os.path.getsize(fname)

    val = size/denom
    print(f"- {fname} is {val} {units.upper()} on disk.")
```

```python
# Getting the size of akas_df in small chunks
get_filesize(fname)
```

### Filtering & Saving akas dataframe

```python
%%time
# Loop through the df_reader itself to get the akas_df
for akas_df in akas_reader:

    ## Replace "\N" with np.nan
    akas_df.replace({'\\N':np.nan}, inplace=True)

    # Keep only US movies
    akas_df = akas_df[(akas_df['region'] == 'US')]

    ## Saving chunk to disk
    fname= f'Data/title_akas_chunk_{chunk_num:03d}.csv.gz'
    akas_df.to_csv(fname, compression='gzip')
    print(f"- Saved {fname}")

    ## Increment chunk_num
    chunk_num+=1
## Closing the reader now that we are done looping through the file
akas_reader.close()
```

```python
# After saving explore all akas_df
akas_df.info()
```

```python
# To load in the all of the saved title akas chunk files & sorted alphabetically
import glob
q = "Data/title_akas_chunk*.csv.gz"
chunked_files = sorted(glob.glob(q))
# Showing the first 5
chunked_files[:5]
```

```python
%%time
## Combining many files
# Loading all files as df and appending to a list
akas_list = []
for file in chunked_files:
    akas_df = pd.read_csv(file, index_col=0, low_memory = False)
    akas_list.append(akas_df)
## Concatenating the list of dfs into 1 combined
akas_combined = pd.concat(akas_list)
akas_combined
```

```python
%%time
## Saving the final combined dataframe
final_fname ='Data/title_akas_combined.csv.gz'
akas_combined.to_csv(final_fname, compression='gzip', index = False)
```

```python
# Getting the memory_usage of df_combined
get_memory_usage(akas_combined)
```

## **Title Basics**

```python
%%time
# Replace "\N" with np.nan
basics.replace({'\\N':np.nan}, inplace=True)
# Eliminate movies that are null for runtimeMinute & genres
basics = basics.dropna(subset=['runtimeMinutes', 'genres'])
basics
```

```python
# Keep only titleType==Movie
basics = basics[(basics['titleType'] == 'movie')]
basics
```

```python
# Exclude movies that are included in the documentary category.
is_documentary = basics['genres'].str.contains('documentary', case=False)
basics = basics[~is_documentary]
basics
```

```python
# Convert startyear to numeric for slicing
basics['startYear'] = basics['startYear'].astype(float).copy()
# keep startYear 2000-2022
basics = basics[(basics['startYear']>=2000)&(basics['startYear']<2022)]
basics
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers = basics['tconst'].isin(akas_df['titleId'])
keepers
```

```python
# Filter basics
basics = basics[keepers]
basics
```

```python
# Check basics dataframe before saving
basics.info()
```

```python
# Save basics dataframe to Data file
basics.to_csv("Data/title_basics.csv.gz", compression='gzip', index = False)
```
